```python
import time, sys
sys.setrecursionlimit(50000)

def fact1(n): #iter loop
	fact = 1
	for i in range(1, n):
		fact = fact * i
	return fact

def fact2(n): #recursion
	if n == 0:
		return 1
	return n * fact2(n - 1)

def compute_e_ver1(n):
	# code for O(n^2)-time version
	fact_list = [1/fact1(x) for x in range(n)]
	result = 0
	for i in range(len(fact_list)):
		result = result + fact_list[i]
	return result

def compute_e_ver2(n):
	result = 1
	k = 1
	for i in range(1,n):
		k = k * i
		result = result + (1/k)
	return result

# n 입력받음
n = int(input())
```

```python
def algo_time(n):  # change parameter to compare different algorithms	
    before1 = time.process_time()
    compute_e_ver1(n)
    after1 = time.process_time()

    before2 = time.process_time()
    compute_e_ver2(n)
    after2 = time.process_time()

    ver_algo = [(after1 - before1), (after2 - before2)]

    return ver_algo

num_list =list(map(lambda x : 2 ** x, range(7,14))) # 2 power n
time_n_ver1 = [] # depend on n
time_n_ver2 = []

for i in num_list: 
	temp = algo_time(i)
	time_n_ver1.append(temp[0])
	time_n_ver2.append(temp[1])

#Plot part
import matplotlib.pyplot as plt
x = num_list
y1 = time_n_ver1
y2 = time_n_ver2
ver1_log_graph= plt.loglog(x, y1) #ver1 graph
ver2_log_graph = plt.loglog(x, y2) #ver2 graph
ver1_big = list(map(lambda i: i**2, x)) # Big-o : O(n^2)
ver2_big = list(map(lambda i: i , x)) #Big-o : O(n)
ver1_big_graph = plt.loglog(x, ver1_big)
ver2_big_graph = plt.loglog(x, ver2_big)
plt.legend( ver1_log_graph+ ver2_log_graph+ ver1_big_graph+ ver2_big_graph,
           ['Ver1 Runtime', 'Ver2 Runtime', 'Ver1 Big-O', 'Ver2 Big-O'])
plt.title('Runtime vs Big-O')
plt.xlabel('n', fontsize = 15)
plt.ylabel('Seconds', fontsize = 15)
plt.show()
```
