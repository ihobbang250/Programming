```python
def gcd_sub(a, b): # a, b gap min -> loop time max 
    while a * b != 0:
        if a > b:
            a = a - b
        else:
            b = b - a
    return a + b

# O(n) = n (a - b = 1, n = a)
gcd_sub(10, 5)
```

```python
def gcd_reminder(a, b): # a, b fibo -> loop time max k time loop
    while a * b != 0:
        if a > b:
            a = a % b
        else:
            b = b % a
    return a + b

# O(n) = log(k) k -> fibo
gcd_reminder(10, 5)
```

```python
#Give next parameter to the recursive 
def gcd_recursive1(a, b): 
    if a * b == 0:
        return a + b
    if a > b : return gcd_recursive1(b, a % b)
    else : return gcd_recursive1(a, b % a)
def gcd_recursive2(a, b): 
    if a * b == 0:
        return a + b
    if a > b : return gcd_recursive2(a - b, b)
    else : return gcd_recursive2(a, b - a)

gcd_recursive1(10, 5)
gcd_recursive2(10, 5)
```

```python
# O(sqrt(n)) algorithm
def number_of_factors(n):
    count, k = 0, 1
    while k * k < n: 
        if n % k == 0:  
            count += 2
            k += 1 # Escape loop
        else: k += 1 # Escape loop
        if k * k == n:
            count += 1
    return count

 
n = int(input())
a = number_of_factors(n)
```

```python
# O(2^n) algorithm
def fibo1(n):
    if n <= 1 : return n
    return fibo1(n - 1) + fibo1(n - 2)

fibo1(10)
```

```python
# O(n) algorithm
def fibo2(n):
    first_num = 0
    second_num = 1
    for i in range(2, n + 1):
        first_num, second_num = second_num, first_num+second_num
    return second_num

fibo2(10)
```

```python
# O(logn) algorithms
import numpy as np

def multi_matrix(arr1, arr2):
    result = np.dot(arr1, arr2).tolist()
    return result

def power(array, n):
    if n == 1: return array
    B = power(array, n//2)
    if n % 2 == 0:
        return multi_matrix(B, B)
    else: return multi_matrix(array, multi_matrix(B, B))

def fibo3(n):
    if n <= 1: return n
    A_n = power([[1, 1], [1, 0]], n)
    return A_n[1][0]

fibo3(10)
```

```python
# O(n) algorithm
# T(n) = T(n/2) + T(n/2 - 1) + c
def fibo4(n):
    if n <= 1: return n
    k = n//2
    F_k = fibo4(k)
    F_k_minus_1 = fibo4(k - 1)
    F_k_plus_1 = F_k_minus_1 + F_k
    
    if n % 2 == 0:
        return F_k*F_k + 2*F_k*F_k_minus_1
    else:
        return F_k_plus_1*F_k_plus_1 + F_k*F_k
    
fibo4(10)
```

```python
def fibo5(n):
    F = [0, 1]
    for i in range(2, n + 1):
        F.append(F[i - 1] + F[i - 2])
    return F[n]

fibo5(10)
```

```python
def quick_select(array, k):
    p = array[0] #pivot
    s = []
    m = []
    l = []
    for i in array:
        if i < p: s.append(i)
        elif i > p: l.append(i)
        else: m.append(i)
    
    if len(s) >= k:
        return quick_select(s, k)
    elif len(s) + len(m) < k:
        return quick_select(l, k - (len(s) + len(m)))
    else:
        return p
```

```python
#MOM algorithms
def find_median_five(array):
    s_array = sorted(array)
    median_index = (len(array)+1) // 2 - 1
    return s_array[median_index]

def MoM(array, k):
    if len(array) == 1: return array[0]
    S, L, M, medians = [], [], [], []
    i = 0
    while i + 4 < len(array):
        medians.append(find_median_five(array[i: i + 5]))
        i = i + 5
    
    if i < len(array) and i + 4 >= len(array):
        medians.append(find_median_five(array[i: ]))
        
    mom = MoM(medians, len(medians)//2)
    
    for v in array:
        if(v < mom): S.append(v)
        elif(v > mom): L.append(v)
        else: M.append(v)

    if(len(S) >= k): return MoM(S, k)
    elif(len(S) + len(M) < k): return MoM(L, k - len(S) - len(M))
    else: return mom

a = [1, 2, 3, 4, 6, 5, 7, 8, 9, 1, 2]
MoM(a, 4)
```

```python
def power1(a, n): # a^n; linear recursion  
    if n == 1:
        return a
    return a * power1(a, n - 1) #T(n) = T(n - 1) + c

def power2(a, n): # a^n; linear recursion ver2
    if n == 0: return 1
    x = power2(a, n/2)
    if n % 2 == 1: 
        return x * x * a
    else: return x * x #T(n) = T(n/2) + c

def power3(a, n): # a^n; linear non-recursion
    x = 1
    for _ in range(n):
        x = x * a
    return x
```

```python
#binary_search algorithm
# array already sorted
def binary_search1(array, k): #slicing recursion
    if len(array) < 1:
        return -1
    m = len(array)//2
    if array[m] < k:
        return binary_search1(array[m+1: ], k)
    elif array[m] > k:
        return binary_search1(array[ :m], k)
    else: return m # T(n) = 1 + n + T(n/2) --> O(n)
    
def binary_search2(array, l, r, k): #non-slicing recursion
    if l > r:
        return -1
    m = (l + r)//2
    if array[m] < k:
        return binary_search2(array, m+1, r, k)
    elif array[m] > k:
        return binary_search2(array, l, m-1, k)
    else: return m  # T(n) = T(n/2) + c --> O(logn)
    
    
def binary_search3(array, k): #non-recursion
    l, r = 0, len(array)
    while l - r <= 0:
        m = (l + r)//2
        if array[m] < k:
            l = m + 1
        elif array[m] > k:
            r = m - 1
        else: return m
        
a = [1, 2, 3, 4, 5]

binary_search3(a, 2)
```

```python
# O(n) algorithm
def linear_rotating(array):
    ex_point = 0
    for i in range(1, len(array)):
        if array[i-1] > array[i]:
            ex_point = i-1
            break
        if i == len(array) - 1: # find nothing
            ex_point = len(array) - 1
    
    rotation_element = len(array) - ex_point - 1 #ex_point is index
    return rotation_element

arr = list(map(int, input().split(" ")))
print(linear_rotating(arr))
```

```python
# O(logn) algorithm
def binary_rotating(array): # find rotating index
    l, r = 0, len(array) - 1 # left index, right index
    check = None # Checking Rotation
    while (l < r) : 
        m = (l + r) // 2 # data size n -> n/2 
        if array[m] < array[l]:
            r = m
            check = True
        elif array[m] > array[r]:
            l = m
            check = True 
        else: break
        
        if (l == m) & (r - m == 1) is True: break
    
    if check is not None:
        return len(array) - (m + 1) # m: rotating index 
    else: return 0 # no rotating case
# T(n) = T(n/2) + c
# O(logn)
arr = list(map(int, input().split(" ")))
print(binary_rotating(arr))
```

```python
def selection_sort(A, n): # A[0] ~ A[n-1]까지 selection sort
    for i in range(n-1, 0, -1): # 큰 수를 차례대로 찾음
        m = get_max_index(A, i) # A[0]~A[i] 중 최대값 인덱스 구함
        A[i],A[m] = A[m],A[i]
    return A

def get_max_index(A, i):
    idx = 0
    for j in range(i+1):
        if A[idx] < A[j] :
            idx = j
    return idx

selection_sort([1, 8, 5, 4], 4)
```

```python
def solve_DP(A, S):
    n = len(A)
    DP = [[0 for s in range(S+1)] for _ in range(n)]
    for i in range(0, n):
        for s in range(1, S+1):
            DP[i][s] = DP[i-1][s]
            if s > A[i]:
                DP[i][s] = DP[i-1][s - A[i]]
            elif s == A[i]:
                DP[i][s] = True
                
    return sum(max(DP))

A = [8, 7, 6, 5, 3, 10, 9]
#S = 15
B = [3, 1, 2, 4, 5]
S = 5
print(solve_DP(B, S))
```
